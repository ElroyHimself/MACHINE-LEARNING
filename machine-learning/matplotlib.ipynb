# matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100)   （0，10，100）前两个参数是起始点 终止点  然后切分100段
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny,color='green')
plt.plot(x,cosy,color='red')
```

```python
plt.plot(x,siny,color='green')
plt.plot(x,cosy,color='red',linestyle='--')  '--'表示虚线
```

```python
plt.plot(x,siny,color='green')
plt.plot(x,cosy,color='red')
plt.xlim(-5,15)    调整x轴左右区间
```

```python
plt.plot(x,siny,color='green')
plt.plot(x,cosy,color='red')
plt.ylim(0,1.5)
```

```python
plt.plot(x,siny,color='green')
plt.plot(x,cosy,color='red')
plt.xlim(-5,15) 
plt.axis([-1,11,-2,2])  前两个是x轴限定条件   后两个是y轴限定条件
```

```python
plt.plot(x,siny,color='green')
plt.plot(x,cosy,color='red')
plt.xlim(-5,15) 
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')
plt.ylabel('y value')   坐标轴加上标签
```

```python
plt.plot(x,siny,color='green',label="sin（x）")
plt.plot(x,cosy,color='red',linestyle='--',label="cos(x))")
plt.xlim(-5,15) 
plt.axis([-1,11,-2,2])
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()  添加图示
```

```python
plt.plot(x,siny,color='green')
plt.plot(x,cosy,color='red')
plt.xlim(-5,15) 
plt.axis([-1,11,-2,2])
plt.title("welcome to machine learning!")
plt.xlabel('x axis')
plt.ylabel('y value')
```
